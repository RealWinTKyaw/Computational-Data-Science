```python
import matplotlib.pyplot as plt
from PIL import Image  # Import Pillow library for reading image files

def visualizationGrid(folder_path, file_path, r, c):

    rows = r
    cols = c
    image_count = 0

    fig, axes = plt.subplots(nrows=rows, ncols=cols, figsize=(10, 10))

    for i in range(rows):
        for j in range(cols):
            if image_count < len(file_path[0:r*c]):
                image_path = os.path.join(folder_path, file_path[image_count])
                image = Image.open(image_path)  # Open image using Pillow
                axes[i, j].imshow(image)
                image_count += 1
```

<b> FAKE DATA VISUALIZED: </b> <br>

```python
fake = r'reduced_dataset\fake\ekhacizpah_mp4'
fake_files = os.listdir(fake)
visualizationGrid(fake, fake_files, 2, 2)
```

<b> CROPPED FAKE DATA : </b> <br>

30 frames extracted from each fake folder 

```python
cropped_fake_consolidated = 'reduced_dataset\cropped_fake_consolidated'
consolidated_fake_files = os.listdir(cropped_fake_consolidated)
visualizationGrid(cropped_fake_consolidated, consolidated_fake_files, 3, 3)
```

<b> CROPPED FAKE DATA AUGMENTED: </b> <br>

Each of the 30 frames from each fake folder augemented into 5 different variations

```python
cropped_fake_augmented = 'reduced_dataset\cropped_fake_augmented'
augmented_fake_files = os.listdir(cropped_fake_augmented)
visualizationGrid(cropped_fake_augmented, augmented_fake_files, 3, 3)
```

<b> REAL DATA VISUALIZED: </b> <br>

```python
real = r'reduced_dataset\real\brwrlczjvi_mp4'
real_files = os.listdir(real)
visualizationGrid(real, real_files, 2, 2)
```

<b> CROPPED REAL DATA : </b> <br>

30 frames extracted from each real folder 

```python
cropped_real_consolidated = 'reduced_dataset\cropped_real_consolidated'
consolidated_real_files = os.listdir(cropped_real_consolidated)
visualizationGrid(cropped_real_consolidated, consolidated_real_files, 3, 3)
```

<b> CROPPED REAL DATA AUGMENTED: </b> <br>

Each of the 30 frames from each real folder augemented into 5 different variations

```python
cropped_real_augmented = 'reduced_dataset\cropped_real_augmented'
augmented_real_files = os.listdir(cropped_real_augmented)
visualizationGrid(cropped_real_augmented, augmented_real_files, 3, 3)
```
